```python
#ceate afunction to find the factorial of a number
```

```python
def fact(n):
    f = 1
    for i in range(1, n + 1):
        f = f * i
    print( f)
```

```python
fact(5)
```

```python
fact(10)
```

```python
def addall(*args):
    s = 0
    for i in args:
        s = s + 1
        return s
```
