```python
text = input("enter a string:")

char1 = input("enter the first character to check: ")
char2 = input("enter the second character to check: ")

found_char1 = False
found_char2 = False

for char in text:
    if char == char1:
        found_char1 = True
    if char == char2:
        found_char2 = True
if found_char1:
    print(f"the character'{char1}'is present in the string.")
else:
    print(f"the character'{char2}'is present in the string.")
if found_char2:  
    print(f"the character '{char2}' is present in the string.)
else:
    print(f"the character '{char2}' is present in the string.")
        
        
```

```python
s = input("Enter a string: ")

count = 0

a = 0
e = 0
i = 0
o = 0
u = 0

for ch in s:
    if ch == 'a' or ch == 'A':
        a += 1
        count += 1

    elif ch == 'e' or ch == 'E':
        e += 1
        count += 1

    elif ch == 'i' or ch == 'I':
        i += 1
        count += 1

    elif ch == 'o' or ch == 'O':
        o += 1
        count += 1

    elif ch == 'u' or ch == 'U':
        u += 1
        count += 1

print("Total vowels =", count)

print("a =", a)
print("e =", e)
print("i =", i)
print("o =", o)
print("u =", u)
```

```python
s = input("Enter a string: ")

rev = ""

for ch in s:
    rev = ch + rev

if s == rev:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
alphabet = set("abcdefghijklmnopqrstuvwxyz")

s = input("Enter string: ").lower()

for ch in s:
    if ch in alphabet:
        alphabet.remove(ch)

print("Missing letters:", alphabet)
```
